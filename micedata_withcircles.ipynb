```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_data(url, header):
    df = pd.read_csv(url, header=header, index_col=0)
    return df
```

```python
filename = 'https://raw.githubusercontent.com/DeepLabCut/DeepLabCut/master/examples/openfield-Pranav-2018-10-30/labeled-data/m4s1/CollectedData_Pranav.csv'
header = [0, 1, 2] ### if multi-animal project, header = [0, 1, 2, 3]

df = read_data(filename, header)
df
```

```python
filenames = df.index
scorer = 'Pranav'
bodypart = 'snout'
n = 5 # 5 pictures with bad labels
```

```python
# importing cv2
import cv2
   
# path
path = r'C:\Users\User\DeepLabCut\examples\openfield-Pranav-2018-10-30\labeled-data\m4s1\img0003.png'
```

```python
# Reading an image in default mode
image = cv2.imread(path)
   
# Window name in which image is displayed
window_name = 'Image'
  
# Center coordinates
center_coordinates1 = (int(df.loc[filenames[3],(scorer, 'rightear', 'x')]), int(df.loc[filenames[3],(scorer, 'rightear', 'y')]))
center_coordinates2 = (int(df.loc[filenames[3],(scorer, 'leftear', 'x')]), int(df.loc[filenames[3],(scorer, 'leftear', 'y')]))
center_coordinates3 = (int(df.loc[filenames[3],(scorer, 'snout', 'x')]), int(df.loc[filenames[3],(scorer, 'snout', 'y')]))

# Radius of circle
radius = 2
  
color1 = (255, 0, 0)
color2 = (0, 255, 0)
color3 = (0, 0, 255)
 
# Line thickness of -1 px
thickness = -1
  
# Using cv2.circle() method
# Draw a circle of red color of thickness -1 px
image = cv2.circle(image, center_coordinates1, radius, color1, thickness)
image = cv2.circle(image, center_coordinates2, radius, color2, thickness)
image = cv2.circle(image, center_coordinates3, radius, color3, thickness)
# Displaying the image
cv2.imshow(window_name, image)
cv2.waitKey()
```
